```python
import requests
import pandas as pd
import matplotlib.pyplot as plt
```

```python
pokemon_list = []

for i in range(1, 21):
    url = f"https://pokeapi.co/api/v2/pokemon/{i}"
    response = requests.get(url)
    data = response.json()
    pokemon_list.append(data)
```

```python
data_json=pokemon_list
data_json[0]
```

```python
df = pd.DataFrame({
    "name": [p["name"] for p in pokemon_list],
    "height": [p["height"] for p in pokemon_list],
    "weight": [p["weight"] for p in pokemon_list],
    "base_experience": [p["base_experience"] for p in pokemon_list],
    "types": [", ".join([t["type"]["name"] for t in p["types"]]) for p in pokemon_list]
})

df.head()
```

```python
df = df[
    [
        "name",
        "types",
        "height",
        "weight",
        "base_experience"
    ]
]

df.head()
```

```python
df = df.rename(columns={
    "name": "pokemon_name",
    "types": "pokemon_types",
    "height": "height_dm",
    "weight": "weight_hg",
    "base_experience": "base_xp"
})

df.head()
```

```python
df = df.sort_values("base_experience", ascending=False)

df
```
